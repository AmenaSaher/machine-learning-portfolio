DataFrames = A tabular data structure with rows and columns. (2 dimensional)

Similar to an Excel spreadsheet

```python
data = {
    "Name":["Misbah","Aishah","Amena"],
    "Age":[12,15,24]
}

df = pd.DataFrame(data, index=["person 1","person 2","person 3"]) #DataFrame constructor

# print(df.loc["person 1"])

# print(df.iloc[1])
# add a new column
df["Job"] = ["Pilot","Doctor","Engineer"]

# add a new rows
new_rows = pd.DataFrame([{"Name":"Ibad","Age":25,"Job":"Engineer"},
                        {"Name":"Mom","Age":49,"Job":"Dentist"}],
                        index = ["person 4","person 5"])

df = pd.concat([df,new_rows])
print(df)
```

Data Cleaning - the process of fixing/removing:

incomplete, incorrect, or irrelevant data. ~75% of work done with Pandas is data cleaning.


```python
# 1. Drop irrelevant columns

# df1 = df1.drop(columns=["Legendary"])

print(df1)
```

```python
# 2. Handle missing data

# df1 = df1.dropna(subset=["Type 2"])
# drop NaN values

df1 = df1.fillna({"Type 2": "None"})
print(df1)
```

```python
# 3. Fix inconsistent values
df1["Type 1"] = df1["Type 1"].replace({"Grass":"GRASS",
                                       "Fire":"FIRE",
                                       "Water":"WATER"})

print(df1)
```

```python
# 4. Standardize text

df1["Name"] = df1["Name"].str.lower()

print(df1.to_string())
```
